```python
import pandas as pd
```

```python
df = pd.read_excel('topic_modeling_doc.xlsx')
```

```python
m = Mystem()
a = list()
i = 0
```

```python
while i!=138:
    lemmas = m.lemmatize(df['text'][i])
    a.append(''.join(lemmas))
    i+=1
```

```python
print(a[1])
```

```python
lem = pd.DataFrame(a)
```

```python
with open('lem.csv', 'w') as outfile:
    lem.to_excel("output.xlsx")
```
